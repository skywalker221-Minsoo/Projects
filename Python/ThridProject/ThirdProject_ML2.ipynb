```python
y = train1['credit']
x = train1.drop(['credit', 'gender', 'car', 'reality', 'income_type',
                'edu_type', 'family_type', 'house_type', 'occyp_type'], axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.1, random_state = 11)
print(X_train.shape, X_test.shape)
```

```python
xgb = XGBClassifier()

xgb_param_grid = {
    'n_estimators' : [100, 200, 400, 600],
    'learning_rate' : [0.01, 0.05, 0.1, 0.15, 0.2],
    'max_depth' : [4, 6, 8, 10, 12],
}

xgb_grid = GridSearchCV(xgb, param_grid = xgb_param_grid, scoring="accuracy", n_jobs=-1, verbose=1)
xgb_grid.fit(X_train, Y_train)
```

```python
print("ìµœê³  í‰ê·  ì •í™•ë„ : {0:.4f}".format(xgb_grid.best_score_))
print("ìµœê³ ì˜ íŒŒë¼ë¯¸í„° : ", xgb_grid.best_params_)
```
